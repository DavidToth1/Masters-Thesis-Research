## This model features a high mass Be star accelerating wind and circumstellar disc, with an elliptical orbit, a 3D coordinate system, and eclipses of a 'pseudo-signal' from a pulsar on the orbit.
## The main goal of this model is to see the number density, dispersion measure, optical depth and spectrum along a line of sight from a point on the orbit given an orbital phase, orbital and disc inclination, eccentricity, argument of periastron and longitude of ascending node.
## Stellar parameters are modelled after the star SS 2883 of the HMXB system PSR B1259-63/SS 2883. Wind terminal velocity is taken from the paper - https://arxiv.org/pdf/0806.4868, while the remaining stellar properties are taken from https://ui.adsabs.harvard.edu/abs/2011ApJ...732L..11N/abstract, https://arxiv.org/abs/1307.7083 and https://arxiv.org/pdf/1804.08402.
## Parameter values of the disc of SS 2883 are taken from the paper https://arxiv.org/abs/2204.08124

# Equations and Geometry
## Electron number density
# $n_{e}(r) = \frac{\rho}{\mu_{e}m_{p}} = \frac{\dot{M}}{4\pi\mu_{e}m_{p}r^2v(r)} = \frac{k}{r^2v(r)}$
Where
# $\rho = \frac{\dot{M}}{4\pi r^2v(r)}$

And

# $k = \frac{\dot{M}}{4\pi\mu_{e}m_{p}}$
With

* $\rho$ =      Density
* $\mu_{e}$ =   Mean molecular weight per free electron (typically set at a value of 1.2)
* $m_{p}$ =     Proton mass
* $\dot{M}$ =   Wind mass loss
* $v(r)$ =     Radial wind velocity
* $r$ =         radial distance from the center of the star

## Accelerating Wind ($\beta$ law)
The wind velocity at a certain distance away from the star is 
# $v(r) = v_0 + (v_{\infty} - v_0)(1-\frac{R_{*}}{r})^\beta$
with 
* $v_0$ =        The wind velocity at stellar surface
* $v_{\infty}$ = Wind terminal velocity
* $R_{*}$ =      Stellar radius
* $\beta$ =      Wind acceleration exponent

By placing the star and orbit at the xy plane origin, we can consider an elliptical orbit in cartesian coordinates
# $\frac{x^2}{a^2} + \frac{y^2}{b^2} = 1$
with
* $a$ = Semi-major axis
* $b$ = Semi-minor axis
The semi-minor axis can be defined as $a\sqrt{1 - e^2}$, where $e^2$ is the eccentricity.
In polar coordinates, since the ellipse is offset by along the x-axis by the focus - $f = ae$, the coordinates relative to the star are
## $x = rcos(\phi) - ae$, $y = bsin(\phi)$, and $r = \sqrt{x^2 + y^2}$
The orbital separation in polar coordinates becomes
# $r = \frac{a(1-e^2)}{1+ecos(\phi)}$
Where $\phi$ is the true anomaly.
The signal position in polar coordinates is defined as $r_s = r(cos(\phi + \omega),sin(\phi + \omega),0)$, where $\omega$ is the argument of periastron. The observer unit vector (direction of signal to observer) is then defined as $\hat{n} = (sin(i)cos(\Omega), sin(i)sin(\Omega), cos(i))$, where $i$ is the inclination and $\Omega$ is the longitude of ascending node. The radial distance of the signal from the star along the path of the line of sight becomes $r(s) = r_s + s\hat{n}$, where $s\ge0$, is the distance along the line of sight. 

The number density at a specific point along the line of sight distance from the orbit can then be described as
# $n_e(r(s)) = \frac{k}{r(s)^2v(r(s))}$

From the definition of $r(s)$, 

## $r(s)^2 = (r_s + s\hat{n})^2 = r^2 + 2rs [cos(\phi + \omega) sin (i)cos(\Omega) + sin(\phi + \omega)sin(i)sin(\Omega)] + s^2$

## Dispersion Measure
# $DM(\phi,i) = \int_{0}^{\infty}n_e(r(s))ds = k\int_{0}^{\infty}\frac{ds}{v(r(s))r(s)^2}$

## Semi-major axis of PSR B1259-63
Given that the orbital period of PSR B1259-63 is approximately 1237 days, its mass is $1.4 M_\odot$ and that the mass of SS 2883 is approximately $20 M_\odot$, using Kepler's 3rd law, the semi-major axis is
# $a^3 = \frac{P^2G(M + m)}{4\pi^2}, a \sim 1345R_\odot$
With
* $P$ = Orbital period
* $G$ = Gravitational constant
* $M$ = Mass of SS 2883
* $m$ = Mass of PSR B1259-63

## Free-free Absorption
Free-free absorption causes signals to be absorbed by the stellar wind depending on its optical depth.
# $\tau_{FFA} \approx 8.24x10^{-2}\nu^{-2.1}T_e^{-1.35}\int n_e^2ds$

with
* $\tau_{FFA}$ = The optical depth
* $\nu$ = The signal frequency 
* $T_e$ = The stellar wind temperature
* $n_e$ = The number density
* $s$ = The line of sight distance

## Flux density
The flux density of the pulsar signal is decribed as
# $S_{\nu} = S_{psr,\nu}e^{-\tau} = \nu^{-p}e^{-\tau}$
with
* $S_{psr,\nu}$ = The intrinsic flux of the pulsar
* $\nu$ = The signal frequency
* $p$ = The power law index
* $\tau$ = The optical depth
When the flux density of the signal is plotted against frequency, it results in the pulsar's spectrum, which changes along eccentric orbits.

## Be Disc
Be stars such as SS 2883 have a circumstellar disc due to the star's immense rotation rate. The disc introduces an additional desnity structure to the system, which is described as
# $\rho(r_{disc}, z) = \rho_0(\frac{r_{disc}}{R_*})^{-\beta}e^{\frac{-z^2}{H(r_{disc})^2}}$
with
* $\rho_0$ = The disc desnity at stellar surface
* $r_{disc}$ = radial distance in the disc plane
* $R_*$ = Stellar radius
* $\beta$ = Disc power law exponent
* $z$ = Height above the disc plane
* $H(r)$ = Scale height

The scale height is described as
# $H(r_{disc}) = c_s\sqrt{\frac{r_{disc}}{GM_*}}r_{disc}$
with
* $c_s$ = Speed of sound
* $G$ = Gravitational constant
* $M_*$ = Stellar mass

And the speed of sound is described as 
# $c_s = \sqrt{\frac{kT}{\mu_em_p}}$
with
* $k$ = Boltzmann constant
* $T$ = Disc temperature

Using the disc desnity profile the number density through the disc becomes
# $n_{e,disc} = \frac{\rho(r_{disc},z)}{\mu_em_p}$

From which the DM and optical depth can be calculated

Placing the disc at the xy plane along with the pulsar orbit allows the radial distance from the pulsar to the observer $r(s)$ serve as the radial distance from the pulsar that passes through the wind to the observer $r(s) = r_{disc}$.

The disc height above its plane is described as $z = r(s)\cdot{\hat{n}_{disc}} = r_s\hat{n}_{disc} + s(\hat{n}\cdot{\hat{n}_{disc}})$. Where an azimuthal component is added to the line of sight unit vector $\hat{n} = (sin(i+\Omega)cos(\Omega), sin(i+\Omega)sin(\Omega), cos(i+\Omega))$, and the disc normal is $\hat{n}_{disc} = (0, sin(i_{disc}), cos(i_{disc}))$, such that the disc rotates on the x-axis.

## Combined Stellar Wind and Be Disc
Since the stellar wind and Be disc do not depend on each-other, combining their number densities is straight forward
# $n_{e,total} = n_{e,wind} + n_{e,disc}$
With this total number density we can calculate the dispersion measure, optical depth, transmission factor and spectrum of the entire system.

## Rotation Measure
The rotation measure shows how much the plane of polarisation of the signal has rotated as a result of traveling through magnetised plasma such as the stellar wind and circumstellar disc. It depends on the number desnity and magnetic field strength along the line of sight. Higher rotation measure occurs when the polarisation rotates more at longer wavelengths, also known as Faraday rotation.

The magnetic field of the star is described as 
# $B = B_0(\frac{R_*}{r(s)})$

With $B_0$, the magnetic field at the stelllar surface assumed to be $0.8 G$ (Johnston et al., 1996 https://ui.adsabs.harvard.edu/abs/1996MNRAS.279.1026J/abstract).

The magnetic field parallel to the line of sight is described as
# $B_{\parallel} = B(\hat{r}\cdot\hat{n})$

Where $\hat{r} = (cos(\phi + \omega), sin(\phi + \omega), 0)$ is the radial unit vector from the star to the signal source, and $\hat{n} = (sin(i)cos(\Omega), sin(i)sin(\Omega), cos(i))$ is the line of sight unit vector from the signal to the observer.

The rotation measure is described as
# $\Delta RM = 8.1x10^5 \int_{LOS} n_eB_{\parallel}ds$ 
(Melatos et al., 1995, https://academic.oup.com/mnras/article/275/2/381/2896129?login=false)

```python
#Packages
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import astropy.units as u
import astropy.constants as con
import ipywidgets as widgets
from IPython.display import display
from scipy.optimize import curve_fit
from scipy.signal import find_peaks
```

```python
#Constants in cgs units
R_sun_cgs = con.R_sun.cgs.value   #Radius of sun in cgs
M_sun_cgs = con.M_sun.cgs.value   #Mass of the sun in cgs
yr_to_sec = u.yr.to(u.s)          #Year to seconds
kms_cgs = (u.km/u.s).to(u.cm/u.s) #Kms in cgs
m_p_cgs = con.m_p.cgs.value       #Mass of proton in cgs
G_cgs = con.G.cgs.value           #Gravitational constant in cgs   
k_B_cgs = con.k_B.cgs.value       #Boltzmann constant in cgs
AU_cgs = (con.au.cgs.value)       #AU in cgs
pc_to_cm = u.pc.to(u.cm)          #PC to cm
```

```python
#Number density function
def num_density(phi, i, a_Rsun, R_star_sol, Mdot_Msun_per_yr, v_0_kms, v_inf_kms, s, e, omega, Omega, beta, mu_e=1.2):
    '''This function calculates the number density at a specific point along an elliptical orbit within a spherically cymmetric wind 
    with accelerating velocity.
    -----------------------------------------------------------------------------------------------------------------------------
    parameters:
    
    phi              [float] -> Orbital Phase (rad)
    i                [float] -> Inclination (rad)
    a_Rsun           [float] -> Orbital radius (solar radii)
    R_star_sol       [float] -> Stellar radius (solar radii)
    Mdot_Msun_per_yr [float] -> Log wind mass loss (solar masses per year)
    v_0_kms          [float] -> Wind velocity at stellar radius (km/s)
    v_inf_kms        [float] -> Terminal velocity of wind (km/s)
    mu_e             [float] -> Mean molecular weight per free electron
    e                [float] -> Eccentricity (0 < e < 1)
    s                [float] -> Distance along line of sight from signal (AU)
    omega            [float] -> Argument of periastron (rad)
    Omega            [float] -> Longitude of ascending node (rad)
    beta             [float] -> Wind acceleration exponent 
    
    -----------------------------------------------------------------------------------------------------------------------------
    Returns:
    
    The function returns the number density [array] -> (cm^-3)'''
    a = a_Rsun * R_sun_cgs                                                                                               #Converts the semi-mmajor axis from solar radii to cm
    R_star = R_star_sol * R_sun_cgs                                                                                      #Converts the stellar radius from solar radii to cm
    Mdot = 10**Mdot_Msun_per_yr * (M_sun_cgs / yr_to_sec)                                                                #Converts the stellar mass loss from solar mass per year to g/s
    v_0 = v_0_kms * kms_cgs                                                                                              #Converts the stellar surface wind velocity ftom km/s to cm/s
    v_inf = v_inf_kms * kms_cgs                                                                                          #Converts the terminal wind velocity from km/s to cm/s
    k = Mdot / (4*np.pi*mu_e*m_p_cgs)                                                                                    #Calculates the k constant (cm^-1)
    r = a*(1 - e**2)/(1 + e*np.cos(phi))                                                                                 #Calculates the orbital separation along the orbit
    r_s_sqrd = r**2 + 2*r*s*(np.cos(phi+omega)*np.sin(i)*np.cos(Omega)+np.sin(phi+omega)*np.sin(i)*np.sin(Omega)) + s**2 #Calculates the squared radial distance from the star along the line of sight
    v_r_s = v_0 + (v_inf-v_0)*(1-R_star/np.sqrt(r_s_sqrd))**beta                                                         #Calculates the velocity of the wind 

    n_e = k/(r_s_sqrd*v_r_s)                                                                                             #Calculates the number density
    n_e[np.sqrt(r_s_sqrd) <= R_star] = 0                                                                                 #Eclipse condition
     
    return n_e
```

```python
#DM function
def dm_beta_wind(phi, i, a_Rsun, R_star_sol, Mdot_Msun_per_yr, v_0_kms, v_inf_kms, e, omega, Omega, beta, mu_e=1.2, s_array_cm=None):
    '''This function calculates the dispersion measure along a line of sight, given orbital phase and inclination, 
    assuming a spherically symmetric wind with an accelerating velocity.
    --------------------------------------------------------------------------------------------------------------
    Parameters:
    
    phi              [float] -> Orbital Phase (rad)
    i                [float] -> Inclination (rad)
    a_Rsun           [float] -> Orbital radius (solar radii)
    R_star_sol       [float] -> Stellar radius (solar radii)
    Mdot_Msun_per_yr [float] -> Log wind mass loss (solar masses per year)
    v_0_kms          [float] -> Wind velocity at stellar radius (km/s)
    v_inf_kms        [float] -> Terminal velocity of wind (km/s)
    e                [float] -> Eccentricity (0 < e < 1)
    omega            [float] -> Argument of periastron (rad)
    Omega            [float] -> Longitude of ascending node (rad)
    beta             [float] -> Wind acceleration exponent 
    mu_e             [float] -> Mean molecular weight per free electron
    
    --------------------------------------------------------------------------------------------------------------
    Returns:
    
    The function returns the dispersion measure [array] -> (pc cm^-3).'''
    a = a_Rsun * R_sun_cgs                                #Converts the semi-major axis from solar radii to cm
    R_star = R_star_sol * R_sun_cgs                       #Converts the stellar radius from solar radii to cm
    Mdot = 10**Mdot_Msun_per_yr * (M_sun_cgs / yr_to_sec) #Converts the stellar mass loss from solar mass per year to g/s
    v_0 = v_0_kms * kms_cgs                               #Converts the stellar surface wind velocity ftom km/s to cm/s
    v_inf = v_inf_kms * kms_cgs                           #Converts the wind velocity from km/s to cm/s
    k = Mdot / (4*np.pi*mu_e*m_p_cgs)                     #Calculates the k constant (cm^-1)

    #Instantiating a line of sight array
    if s_array_cm is None:
        s_array_cm = np.logspace(np.log10(1e5), np.log10(12 * AU_cgs), 512)  
    s = np.asarray(s_array_cm, dtype=float)

    #Initialising a DM array
    DM_array = []

    for value in phi:
        r = a * (1 - e**2) / (1 + e * np.cos(value))                                 #Calculates the orbital separation along the orbit
        r_s_sqrd = r**2 + 2*r*s*(np.cos(value+omega)*np.sin(i)*np.cos(Omega) +       #Calculates the squared radial distance from the star along the line of sight
                                 np.sin(value+omega)*np.sin(i)*np.sin(Omega)) + s**2
        v_r_s = v_0 + (v_inf - v_0) * (1 - R_star/np.sqrt(r_s_sqrd))**beta           #Calculates the velocity of the wind 
        n_e = k / (r_s_sqrd * v_r_s)                                                 #Calculates the number density
        n_e[np.sqrt(r_s_sqrd) <= R_star] = 0                                         #Eclipse condition

        DM_cgs = np.trapz(n_e, s)                                                    #Integrating the DM in cm^-2
        DM = (DM_cgs / pc_to_cm)                                                     #Converting DM to pc cm^3
        DM_array.append(DM)

    return np.array(DM_array) 
```

```python
#Free-free absorption function
def ffa(phi, i, a_Rsun, R_star_sol, Mdot_Msun_per_yr, v_0_kms, v_inf_kms, e, omega, Omega, beta, nu, T, mu_e=1.2, s_array_cm=None):
    '''This function calculates the free-free absorption along a line of sight, given orbital phase and inclination, 
    assuming a spherically symmetric wind with an accelerating velocity.
    --------------------------------------------------------------------------------------------------------------
    Parameters:
    
    phi              [float] -> Orbital Phase (rad)
    i                [float] -> Inclination (rad)
    a_Rsun           [float] -> Orbital radius (solar radii)
    R_star_sol       [float] -> Stellar radius (solar radii)
    Mdot_Msun_per_yr [float] -> Log wind mass loss (solar masses per year)
    v_0_kms          [float] -> Wind velocity at stellar radius (km/s)
    v_inf_kms        [float] -> Terminal velocity of wind (km/s)
    e                [float] -> Eccentricity (0 < e < 1)
    omega            [float] -> Argument of periastron (rad)
    Omega            [float] -> Longitude of ascending node (rad)
    beta             [float] -> Wind acceleration exponent
    nu               [float] -> Signal frequency (GHz)
    T                [float] -> Stellar wind temperature (K)
    mu_e             [float] -> Mean molecular weight per free electron
    
    --------------------------------------------------------------------------------------------------------------
    Returns:
    
    The function returns the optical depth [array].'''
    a = a_Rsun * R_sun_cgs                                #Converts the semi-major axis from solar radii to cm
    R_star = R_star_sol * R_sun_cgs                       #Converts the stellar radius from solar radii to cm
    Mdot = 10**Mdot_Msun_per_yr * (M_sun_cgs / yr_to_sec) #Converts the stellar mass loss from solar mass per year to g/s
    v_0 = v_0_kms * kms_cgs                               #Converts the stellar surface wind velocity ftom km/s to cm/s
    v_inf = v_inf_kms * kms_cgs                           #Converts the wind velocity from km/s to cm/s
    k = Mdot / (4*np.pi*mu_e*m_p_cgs)                     #Calculates the k constant (cm^-1)

    #Instantiating a line of sight array
    if s_array_cm is None:
        s_array_cm = np.logspace(np.log10(1e5), np.log10(12 * AU_cgs), 512)  
    s = np.asarray(s_array_cm, dtype=float)

    #Initialising an optical depth array
    tau_array = []

    for value in phi:
        r = a * (1 - e**2) / (1 + e * np.cos(value))                                 #Calculates the orbital separation along the orbit
        r_s_sqrd = r**2 + 2*r*s*(np.cos(value+omega)*np.sin(i)*np.cos(Omega) +       #Calculates the squared radial distance from the star along the line of sight
                                 np.sin(value+omega)*np.sin(i)*np.sin(Omega)) + s**2
        v_r_s = v_0 + (v_inf - v_0) * (1 - R_star/np.sqrt(r_s_sqrd))**beta           #Calculates the velocity of the wind 
        n_e = k / (r_s_sqrd * v_r_s)                                                 #Calculates the number density
        n_e[np.sqrt(r_s_sqrd) <= R_star] = 0                                         #Eclipse condition

        tau = 8.24e-2*nu**-2.1*T**-1.35*np.trapz(n_e**2, s / pc_to_cm)               #Integrating the optical depth
        tau_array.append(tau)

    return np.array(tau_array) 
```

```python
#Flux density function
def spectrum(phi, i, a_Rsun, R_star_sol, Mdot_Msun_per_yr, v_0_kms, v_inf_kms, e, omega, Omega, beta, nu, T, p, mu_e=1.2, s_array_cm=None):
    '''This function calculates the spectrum of the pulsar for a given inclination and frequency along the orbit.
    -------------------------------------------------------------------------------------------------------------
    parameters:
    
    phi              [float] -> Orbital Phase (rad)
    i                [float] -> Inclination (rad)
    a_Rsun           [float] -> Orbital radius (solar radii)
    R_star_sol       [float] -> Stellar radius (solar radii)
    Mdot_Msun_per_yr [float] -> Log wind mass loss (solar masses per year)
    v_0_kms          [float] -> Wind velocity at stellar radius (km/s)
    v_inf_kms        [float] -> Terminal velocity of wind (km/s)
    e                [float] -> Eccentricity (0 < e < 1)
    omega            [float] -> Argument of periastron (rad)
    Omega            [float] -> Longitude of ascending node (rad)
    beta             [float] -> Wind acceleration exponent
    nu               [float] -> Signal frequency (GHz)
    T                [float] -> Stellar wind temperature (K)
    p                [float] -> Power law exponent
    mu_e             [float] -> Mean molecular weight per free electron
    
    -------------------------------------------------------------------------------------------------------------
    Returns:
    
    The function returns the pulsar spectrum [array] (Jy)'''
    a = a_Rsun * R_sun_cgs                                #Converts the semi-major axis from solar radii to cm
    R_star = R_star_sol * R_sun_cgs                       #Converts the stellar radius from solar radii to cm
    Mdot = 10**Mdot_Msun_per_yr * (M_sun_cgs / yr_to_sec) #Converts the stellar mass loss from solar mass per year to g/s
    v_0 = v_0_kms * kms_cgs                               #Converts the stellar surface wind velocity ftom km/s to cm/s
    v_inf = v_inf_kms * kms_cgs                           #Converts the wind velocity from km/s to cm/s
    k = Mdot / (4*np.pi*mu_e*m_p_cgs)                     #Calculates the k constant (cm^-1)

    #Instantiating a line of sight array
    if s_array_cm is None:
        s_array_cm = np.logspace(np.log10(1e5), np.log10(12 * AU_cgs), 512)  
    s = np.asarray(s_array_cm, dtype=float)

    #Initialising a flux density array
    S_nu_array = []

    for value in phi:
        for freq in nu:
            r = a * (1 - e**2) / (1 + e * np.cos(value))                                 #Calculates the orbital separation along the orbit
            r_s_sqrd = r**2 + 2*r*s*(np.cos(value+omega)*np.sin(i)*np.cos(Omega) +       #Calculates the squared radial distance from the star along the line of sight
                                     np.sin(value+omega)*np.sin(i)*np.sin(Omega)) + s**2
            v_r_s = v_0 + (v_inf - v_0) * (1 - R_star/np.sqrt(r_s_sqrd))**beta           #Calculates the velocity of the wind 
            n_e = k / (r_s_sqrd * v_r_s)                                                 #Calculates the number density
            n_e[np.sqrt(r_s_sqrd) <= R_star] = 0                                         #Eclipse condition

            tau = 8.24e-2*freq**-2.1*T**-1.35*np.trapz(n_e**2, s / pc_to_cm)             #Integrating the optical depth
            S_nu = freq**(-p)*np.exp(-tau)                                               #Calculating the flux density
            S_nu_array.append(S_nu)
    S_nu_2D = np.array(S_nu_array).reshape(len(phi), len(nu))                            #Reshaping to a 2D array
    return S_nu_2D
```

```python
#Disc number density function
def disc_ne(a_Rsun, T, M_star_sol, R_star_sol, beta, rho0, i_disc, phi, omega, Omega, e, i, s, mu_e=1.2):
    '''This function calculates the number density along the line of sight from the
    pulsar through the disc for a given inclination and an eccentric orbit.
    -------------------------------------------------------------------------------
    parameters:
    
    a_Rsun     [float] -> Orbital radius (solar radii)
    T          [float] -> Be disc temperature (K)
    M_star_sol [float] -> Stellar mass (solar masses)
    R_star_sol [float] -> Stellar radius (solar radii)
    beta       [float] -> Disc power law exponent
    rho0       [float] -> Log disc density at stellar surface (gcm^-3)
    i_disc     [float] -> Inclination of disc plane (rad)
    phi        [float] -> Orbital Phase (rad)
    omega      [float] -> Argument of periastron (rad)
    Omega      [float] -> Longitude of ascending node (rad)
    e          [float] -> Eccentricity of orbit
    i          [float] -> Orbital Inclination
    s          [array] -> Array of line of sight points (cm)
    mu_e       [float] -> Mean molecular weight per free electron
    
    --------------------------------------------------------------------------------
    Returns:
    
    The function returns the number desnity [array] (cm^-3)'''
    a = a_Rsun * R_sun_cgs               #Converts the orbital radius from solar radii to cm
    R_star = R_star_sol * R_sun_cgs      #Converts the stellar radius from solar radii to cm
    M_star = M_star_sol * M_sun_cgs      #Converts the stellar mass from solar mass to g
    r = a*(1 - e**2)/(1 + e*np.cos(phi)) #Calculates the orbital separation along the orbit

    r_disc = np.sqrt(r**2 + 2*r*s*(np.cos(phi+omega)*np.sin(i)*np.cos(Omega)+np.sin(phi+omega)*np.sin(i)*np.sin(Omega)) + s**2) #Projected radius through disc plane   
    c_s = np.sqrt((k_B_cgs * T)/(mu_e * m_p_cgs))                                                                               #Calculates the disc speed of sound
    H_rdisc = c_s * np.sqrt(r_disc/(G_cgs * M_star)) * r_disc                                                                   #Calculates the scale height
    z = r*np.sin(phi+omega)*np.sin(i_disc) + s*(np.sin(i+Omega)*np.sin(Omega)*np.sin(i_disc) + np.cos(i+Omega)*np.cos(i_disc))  #Calculates the height above the disc plane
    rho_disc = 10**rho0 * (r_disc/R_star)**(-beta) * np.exp(-z**2/H_rdisc**2)                                                   #Calculates the disc desnity profile
    ne_disc = rho_disc/(mu_e * m_p_cgs)                                                                                         #Calculates the number density
    ne_disc[r_disc < R_star] = 0                                                                                                #Constraining the inner disc radius

    return ne_disc
```

```python
#Disc dispersion measure function
def disc_DM(a_Rsun, T, M_star_sol, R_star_sol, beta, rho0, i_disc, phi, omega, Omega, e, i, mu_e=1.2, s_array_cm=None):
    '''This function calculates dispersion measure along 
    the line of sight from the pulsar through the disc for a given inclination and an eccentric orbit.
    -------------------------------------------------------------------------------
    parameters:
    
    a_Rsun     [float] -> Orbital radius (solar radii)
    T          [float] -> Be disc temperature (K)
    M_star_sol [float] -> Stellar mass (solar masses)
    R_star_sol [float] -> Stellar radius (solar radii)
    beta       [float] -> Disc power law exponent
    rho0       [float] -> Log disc density at stellar surface (gcm^-3)
    i_disc     [float] -> Inclination of disc plane (rad)
    phi        [float] -> Orbital Phase (rad)
    omega      [float] -> Orbit phase angle (rad)
    Omega      [float] -> Longitude of ascending node (rad)
    e          [float] -> Eccentricity of orbit
    i          [float] -> Orbital Inclination
    mu_e       [float] -> Mean molecular weight per free electron
    
    --------------------------------------------------------------------------------
    Returns:
    
    The function returns the dispersion measure [array] (pc cm^-3)'''
    a = a_Rsun * R_sun_cgs          #Converts the orbital radius from solar radii to cm
    R_star = R_star_sol * R_sun_cgs #Converts the stellar radius from solar radii to cm
    M_star = M_star_sol * M_sun_cgs #Converts the stellar mass from solar mass to g

    #Instantiating a line of sight array
    if s_array_cm is None:
        s_array_cm = np.logspace(np.log10(1e5), np.log10(12 * AU_cgs), 512)  
    s = np.asarray(s_array_cm, dtype=float)

    #Initialising a DM array
    disc_DM_array = []

    for value in phi:
        r = a*(1 - e**2)/(1 + e*np.cos(value))                                                                                          #Calculates the orbital separation along the orbit
        r_disc = np.sqrt(r**2 + 2*r*s*(np.cos(value+omega)*np.sin(i)*np.cos(Omega)+np.sin(value+omega)*np.sin(i)*np.sin(Omega)) + s**2) #Projected radius through disc plane        
        c_s = np.sqrt((k_B_cgs * T)/(mu_e * m_p_cgs))                                                                                   #Calculates the disc speed of sound
        H_rdisc = c_s * np.sqrt(r_disc/(G_cgs * M_star)) * r_disc                                                                       #Calculates the scale height
        z = r*np.sin(value+omega)*np.sin(i_disc) + s*(np.sin(i+Omega)*np.sin(Omega)*np.sin(i_disc) + np.cos(i+Omega)*np.cos(i_disc))      #Calculates the height above the disc plane
        rho_disc = 10**rho0 * (r_disc/R_star)**(-beta) * np.exp(-z**2/H_rdisc**2)                                                       #Calculates the disc desnity profile
        ne_disc = rho_disc/(mu_e * m_p_cgs)                                                                                             #Calculates the number density
        ne_disc[r_disc < R_star] = 0                                                                                                    #Constraining the inner disc radius


        disc_DM_cgs = np.trapz(ne_disc, s)                                                                                              #Integrating the DM in cm^-2
        disc_DM = (disc_DM_cgs / pc_to_cm)                                                                                              #Converting DM to pc cm^3
        disc_DM_array.append(disc_DM)

    return np.array(disc_DM_array) 
```

```python
#Disc free-free absorption function
def disc_ffa(a_Rsun, T, M_star_sol, R_star_sol, beta, rho0, i_disc, phi, omega, Omega, e, i, nu, mu_e=1.2, s_array_cm=None):
    '''This function calculates optical depth along 
    the line of sight from the pulsar through the disc for a given inclination and an eccentric orbit.
    -------------------------------------------------------------------------------
    parameters:
    
    a_Rsun     [float] -> Orbital radius (solar radii)
    T          [float] -> Be disc temperature (K)
    M_star_sol [float] -> Stellar mass (solar masses)
    R_star_sol [float] -> Stellar radius (solar radii)
    beta       [float] -> Disc power law exponent
    rho0       [float] -> Log disc density at stellar surface (gcm^-3)
    i_disc     [float] -> Inclination of disc plane (rad)
    phi        [float] -> Orbital Phase (rad)
    omega      [float] -> Orbit phase angle (rad)
    Omega      [float] -> Longitude of ascending node (rad)
    e          [float] -> Eccentricity of orbit
    i          [float] -> Orbital Inclination
    nu         [float] -> Signal frequency (GHz)
    mu_e       [float] -> Mean molecular weight per free electron
    
    --------------------------------------------------------------------------------
    Returns:
    
    The function returns the optical depth [array]'''
    a = a_Rsun * R_sun_cgs          #Converts the orbital radius from solar radii to cm
    R_star = R_star_sol * R_sun_cgs #Converts the stellar radius from solar radii to cm
    M_star = M_star_sol * M_sun_cgs #Converts the stellar mass from solar mass to g

    #Instantiating a line of sight array
    if s_array_cm is None:
        s_array_cm = np.logspace(np.log10(1e5), np.log10(12 * AU_cgs), 512)  
    s = np.asarray(s_array_cm, dtype=float)

    #Initialising an optical depth array
    disc_tau_array = []
    
    for value in phi:
        r = a*(1 - e**2)/(1 + e*np.cos(value))                                                                                          #Calculates the orbital separation along the orbit
        r_disc = np.sqrt(r**2 + 2*r*s*(np.cos(value+omega)*np.sin(i)*np.cos(Omega)+np.sin(value+omega)*np.sin(i)*np.sin(Omega)) + s**2) #Projected radius through disc plane         
        c_s = np.sqrt((k_B_cgs * T)/(mu_e * m_p_cgs))                                                                                   #Calculates the disc speed of sound
        H_rdisc = c_s * np.sqrt(r_disc/(G_cgs * M_star)) * r_disc                                                                       #Calculates the scale height
        z = r*np.sin(value+omega)*np.sin(i_disc) + s*(np.sin(i+Omega)*np.sin(Omega)*np.sin(i_disc) + np.cos(i+Omega)*np.cos(i_disc))      #Calculates the height above the disc plane
        rho_disc = 10**rho0 * (r_disc/R_star)**(-beta) * np.exp(-z**2/H_rdisc**2)                                                       #Calculates the disc desnity profile
        ne_disc = rho_disc/(mu_e * m_p_cgs)                                                                                             #Calculates the number density
        ne_disc[r_disc < R_star] = 0                                                                                                    #Constraining the inner disc radius


        disc_tau = 8.24e-2*nu**(-2.1)*T**(-1.35)*np.trapz(ne_disc**2, s / pc_to_cm)                                                     #Integrating tau
        disc_tau_array.append(disc_tau)

    return np.array(disc_tau_array) 
```

```python
#Disc flux density function
def disc_spectrum(a_Rsun, T, M_star_sol, R_star_sol, beta, rho0, i_disc, phi, omega, Omega, e, i, nu, p, mu_e=1.2, s_array_cm=None):
    '''This function calculates optical depth along 
    the line of sight from the pulsar through the disc for a given inclination and an eccentric orbit.
    -------------------------------------------------------------------------------
    parameters:
    
    a_Rsun     [float] -> Orbital radius (solar radii)
    T          [float] -> Be disc temperature (K)
    M_star_sol [float] -> Stellar mass (solar masses)
    R_star_sol [float] -> Stellar radius (solar radii)
    beta       [float] -> Disc power law exponent
    rho0       [float] -> Log disc density at stellar surface (gcm^-3)
    i_disc     [float] -> Inclination of disc plane (rad)
    phi        [float] -> Orbital Phase (rad)
    omega      [float] -> Orbit phase angle (rad)
    Omega      [float] -> Longitude of ascending node (rad)
    e          [float] -> Eccentricity of orbit
    i          [float] -> Orbital Inclination
    nu         [float] -> Signal frequency (GHz)
    p          [float] -> Power law exponent
    mu_e       [float] -> Mean molecular weight per free electron
    
    --------------------------------------------------------------------------------
    Returns:
    
    The function returns the optical depth [array]'''
    a = a_Rsun * R_sun_cgs          #Converts the orbital radius from solar radii to cm
    R_star = R_star_sol * R_sun_cgs #Converts the stellar radius from solar radii to cm
    M_star = M_star_sol * M_sun_cgs #Converts the stellar mass from solar mass to g
    
    #Instantiating a line of sight array
    if s_array_cm is None:
        s_array_cm = np.logspace(np.log10(1e5), np.log10(12 * AU_cgs), 512)  
    s = np.asarray(s_array_cm, dtype=float)

    #Initialising an optical depth array
    disc_S_nu_array = []
    
    for value in phi:
        for freq in nu:
            r = a*(1 - e**2)/(1 + e*np.cos(value))                                                                                          #Calculates the orbital separation along the orbit
            r_disc = np.sqrt(r**2 + 2*r*s*(np.cos(value+omega)*np.sin(i)*np.cos(Omega)+np.sin(value+omega)*np.sin(i)*np.sin(Omega)) + s**2) #Projected radius through disc plane             
            c_s = np.sqrt((k_B_cgs * T)/(mu_e * m_p_cgs))                                                                                   #Calculates the disc speed of sound
            H_rdisc = c_s * np.sqrt(r_disc/(G_cgs * M_star)) * r_disc                                                                       #Calculates the scale height
            z = r*np.sin(value+omega)*np.sin(i_disc) + s*(np.sin(i+Omega)*np.sin(Omega)*np.sin(i_disc) + np.cos(i+Omega)*np.cos(i_disc))      #Calculates the height above the disc plane
            rho_disc = 10**rho0 * (r_disc/R_star)**(-beta) * np.exp(-z**2/H_rdisc**2)                                                       #Calculates the disc desnity profile
            ne_disc = rho_disc/(mu_e * m_p_cgs)                                                                                             #Calculates the number density
            ne_disc[r_disc < R_star] = 0                                                                                                    #Constraining the inner disc radius


            disc_tau = 8.24e-2*freq**(-2.1)*T**(-1.35)*np.trapz(ne_disc**2, s / pc_to_cm)                                                   #Integrating tau
            disc_S_nu = freq**(-p)*np.exp(-disc_tau)                                                                                        #Calculatinng the flux density
            disc_S_nu_array.append(disc_S_nu)
    disc_S_nu_2D = np.array(disc_S_nu_array).reshape(len(phi), len(nu))                                                                     #Reshaping to a 2D array
    return np.array(disc_S_nu_2D) 
```

```python
#Total wind-disc number density function
def tot_num_density(phi, i, a_Rsun, R_star_sol, M_star_sol, Mdot_Msun_per_yr, v_0_kms, v_inf_kms, s, e, omega, Omega, beta, T_disc, beta_disc, rho0, i_disc, mu_e=1.2):
    '''This function calculates the total number density at a specific point along an elliptical orbit within a spherically cymmetric wind 
    with accelerating velocity and an axisymmetric disc.
    -----------------------------------------------------------------------------------------------------------------------------
    parameters:
    
    phi              [float] -> Orbital Phase (rad)
    i                [float] -> Inclination (rad)
    a_Rsun           [float] -> Orbital radius (solar radii)
    R_star_sol       [float] -> Stellar radius (solar radii)
    M_star_sol       [float] -> Stellar mass (solar masses)
    Mdot_Msun_per_yr [float] -> Log wind mass loss (solar masses per year)
    v_0_kms          [float] -> Wind velocity at stellar radius (km/s)
    v_inf_kms        [float] -> Terminal velocity of wind (km/s)
    s                [float] -> Distance along line of sight from signal (AU)
    e                [float] -> Eccentricity (0 < e < 1)
    omega            [float] -> Argument of periastron (rad)
    Omega            [float] -> Longitude of ascending node (rad)
    beta             [float] -> Wind acceleration exponent
    T_disc           [float] -> Temperature of disc (K)
    beta_disc        [floar] -> Disc power-law exponent
    rho0             [float] -> Log disc density at stellar surface (gcm^-3)
    i_disc           [float] -> Inclination of disc
    mu_e             [float] -> Mean molecular weight per free electron
    
    -----------------------------------------------------------------------------------------------------------------------------
    Returns:
    
    The function returns the total number density [array] -> (cm^-3)'''
    #Stellar wind number density
    a = a_Rsun * R_sun_cgs                                                                                               #Converts the semi-mmajor axis from solar radii to cm
    R_star = R_star_sol * R_sun_cgs                                                                                      #Converts the stellar radius from solar radii to cm
    Mdot = 10**Mdot_Msun_per_yr * (M_sun_cgs / yr_to_sec)                                                                #Converts the stellar mass loss from solar mass per year to g/s
    v_0 = v_0_kms * kms_cgs                                                                                              #Converts the stellar surface wind velocity ftom km/s to cm/s
    v_inf = v_inf_kms * kms_cgs                                                                                          #Converts the terminal wind velocity from km/s to cm/s
    k = Mdot / (4*np.pi*mu_e*m_p_cgs)                                                                                    #Calculates the k constant (cm^-1)
    r = a*(1 - e**2)/(1 + e*np.cos(phi))                                                                                 #Calculates the orbital separation along the orbit
    r_s_sqrd = r**2 + 2*r*s*(np.cos(phi+omega)*np.sin(i)*np.cos(Omega)+np.sin(phi+omega)*np.sin(i)*np.sin(Omega)) + s**2 #Calculates the squared radial distance from the star along the line of sight
    v_r_s = v_0 + (v_inf-v_0)*(1-R_star/np.sqrt(r_s_sqrd))**beta                                                         #Calculates the velocity of the wind 
    n_e = k/(r_s_sqrd*v_r_s)                                                                                             #Calculates the number density
    n_e[np.sqrt(r_s_sqrd) <= R_star] = 0                                                                                 #Eclipse condition
    
    #Disc number density
    M_star = M_star_sol * M_sun_cgs                                                                                            #Converts the stellar mass from solar mass to g
    r_disc = np.sqrt(r_s_sqrd)                                                                                                 #Projected radius through disc plane   
    c_s = np.sqrt((k_B_cgs * T_disc)/(mu_e * m_p_cgs))                                                                         #Calculates the disc speed of sound
    H_rdisc = c_s * np.sqrt(r_disc/(G_cgs * M_star)) * r_disc                                                                  #Calculates the scale height
    z = r*np.sin(phi+omega)*np.sin(i_disc) + s*(np.sin(i+Omega)*np.sin(Omega)*np.sin(i_disc) + np.cos(i+Omega)*np.cos(i_disc)) #Calculates the height above the disc plane
    rho_disc = 10**rho0 * (r_disc/R_star)**(-beta_disc) * np.exp(-z**2/H_rdisc**2)                                             #Calculates the disc desnity profile
    ne_disc = rho_disc/(mu_e * m_p_cgs)                                                                                        #Calculates the number density
    ne_disc[r_disc < R_star] = 0                                                                                               #Constraining the inner disc radius

    #Total number density
    total_ne = n_e + ne_disc

    return total_ne
```

```python
#Total wind-disc free-free absorption function
def tot_ffa(phi, i, a_Rsun, R_star_sol, M_star_sol, Mdot_Msun_per_yr, v_0_kms, v_inf_kms, e, omega, Omega, beta, T_wind, T_disc, beta_disc, rho0, i_disc, nu, mu_e=1.2, s_array_cm=None):
    '''This function calculates the optical depth along 
    the line of sight from the pulsar through the disc and stellar wind for a given inclination and an eccentric orbit.
    -----------------------------------------------------------------------------------------------------------------------------
    parameters:
    
    phi              [float] -> Orbital Phase (rad)
    i                [float] -> Inclination (rad)
    a_Rsun           [float] -> Orbital radius (solar radii)
    R_star_sol       [float] -> Stellar radius (solar radii)
    M_star_sol       [float] -> Stellar mass (solar masses)
    Mdot_Msun_per_yr [float] -> Log wind mass loss (solar masses per year)
    v_0_kms          [float] -> Wind velocity at stellar radius (km/s)
    v_inf_kms        [float] -> Terminal velocity of wind (km/s)
    e                [float] -> Eccentricity (0 < e < 1)
    omega            [float] -> Argument of periastron (rad)
    Omega            [float] -> Longitude of ascending node (rad)
    beta             [float] -> Wind acceleration exponent
    T_disc           [float] -> Temperature of disc (K)
    T_wind           [float] -> Temperature of wind (K)
    beta_disc        [floar] -> Disc power-law exponent
    rho0             [float] -> Log disc density at stellar surface (gcm^-3)
    i_disc           [float] -> Inclination of disc
    nu               [float] -> Signal frequency (GHz)
    mu_e             [float] -> Mean molecular weight per free electron
    s_array_cm       [array] -> Distance along line of sight from signal (AU)
    
    -----------------------------------------------------------------------------------------------------------------------------
    Returns:
    
    The function returns the total optical depth [array] -> (cm^-3)'''
    #Instantiating a line of sight array
    if s_array_cm is None:
        s_array_cm = np.logspace(np.log10(1e5), np.log10(12 * AU_cgs), 512)  
    s = np.asarray(s_array_cm, dtype=float)

    #Initialising a DM array
    tot_tau_array = []
  
    a = a_Rsun * R_sun_cgs                                #Converts the semi-mmajor axis from solar radii to cm
    R_star = R_star_sol * R_sun_cgs                       #Converts the stellar radius from solar radii to cm
    Mdot = 10**Mdot_Msun_per_yr * (M_sun_cgs / yr_to_sec) #Converts the stellar mass loss from solar mass per year to g/s
    v_0 = v_0_kms * kms_cgs                               #Converts the stellar surface wind velocity ftom km/s to cm/s
    v_inf = v_inf_kms * kms_cgs                           #Converts the terminal wind velocity from km/s to cm/s
    k = Mdot / (4*np.pi*mu_e*m_p_cgs)                     #Calculates the k constant (cm^-1)
    M_star = M_star_sol * M_sun_cgs                       #Converts the stellar mass from solar mass to g

    for value in phi:
        #Stellar wind number density
        r = a*(1 - e**2)/(1 + e*np.cos(value))                                                                                   #Calculates the orbital separation along the orbit
        r_s_sqrd = r**2 + 2*r*s*(np.cos(value+omega)*np.sin(i)*np.cos(Omega)+np.sin(value+omega)*np.sin(i)*np.sin(Omega)) + s**2 #Calculates the squared radial distance from the star along the line of sight
        v_r_s = v_0 + (v_inf-v_0)*(1-R_star/np.sqrt(r_s_sqrd))**beta                                                             #Calculates the velocity of the wind 
        n_e = k/(r_s_sqrd*v_r_s)                                                                                                 #Calculates the number density
        n_e[np.sqrt(r_s_sqrd) <= R_star] = 0                                                                                     #Eclipse condition
    
        #Disc number density
        r_disc = np.sqrt(r_s_sqrd)                                                                                                 #Projected radius through disc plane   
        c_s = np.sqrt((k_B_cgs * T_disc)/(mu_e * m_p_cgs))                                                                         #Calculates the disc speed of sound
        H_rdisc = c_s * np.sqrt(r_disc/(G_cgs * M_star)) * r_disc                                                                  #Calculates the scale height
        z = r*np.sin(value+omega)*np.sin(i_disc) + s*(np.sin(i+Omega)*np.sin(Omega)*np.sin(i_disc) + np.cos(i+Omega)*np.cos(i_disc)) #Calculates the height above the disc plane
        rho_disc = 10**rho0 * (r_disc/R_star)**(-beta_disc) * np.exp(-z**2/H_rdisc**2)                                             #Calculates the disc desnity profile
        ne_disc = rho_disc/(mu_e * m_p_cgs)                                                                                        #Calculates the number density
        ne_disc[r_disc < R_star] = 0                                                                                               #Constraining the inner disc radius

        #Total number density
        total_ne = n_e + ne_disc

        total_tau = 8.24e-2*nu**(-2.1)*((T_wind+T_disc)/2)**(-1.35)*np.trapz(total_ne**2, s / pc_to_cm) #Integrating the optical depth
        tot_tau_array.append(total_tau)

    return np.array(tot_tau_array) 
```

```python
#Total wind-disc flux density function
def tot_spectrum(phi, i, a_Rsun, R_star_sol, M_star_sol, Mdot_Msun_per_yr, v_0_kms, v_inf_kms, e, omega, Omega, beta, T_wind, T_disc, beta_disc, rho0, i_disc, nu, p, mu_e=1.2, s_array_cm=None):
    '''This function calculates the flux density along 
    the line of sight from the pulsar through the disc and stellar wind for a given inclination and an eccentric orbit.
    -----------------------------------------------------------------------------------------------------------------------------
    parameters:
    
    phi              [float] -> Orbital Phase (rad)
    i                [float] -> Inclination (rad)
    a_Rsun           [float] -> Orbital radius (solar radii)
    R_star_sol       [float] -> Stellar radius (solar radii)
    M_star_sol       [float] -> Stellar mass (solar masses)
    Mdot_Msun_per_yr [float] -> Log wind mass loss (solar masses per year)
    v_0_kms          [float] -> Wind velocity at stellar radius (km/s)
    v_inf_kms        [float] -> Terminal velocity of wind (km/s)
    e                [float] -> Eccentricity (0 < e < 1)
    omega            [float] -> Argument of periastron (rad)
    Omega            [float] -> Longitude of ascending node (rad)
    beta             [float] -> Wind acceleration exponent
    T_disc           [float] -> Temperature of disc (K)
    T_wind           [float] -> Temperature of wind (K)
    beta_disc        [floar] -> Disc power-law exponent
    rho0             [float] -> Log disc density at stellar surface (gcm^-3)
    i_disc           [float] -> Inclination of disc
    nu               [float] -> Signal frequency (GHz)
    p                [float] -> Spectrum power law exponent
    mu_e             [float] -> Mean molecular weight per free electron
    s_array_cm       [array] -> Distance along line of sight from signal (AU)
    
    -----------------------------------------------------------------------------------------------------------------------------
    Returns:
    
    The function returns the total flux density [array] -> (cm^-3)'''
    #Instantiating a line of sight array
    if s_array_cm is None:
        s_array_cm = np.logspace(np.log10(1e5), np.log10(12 * AU_cgs), 512)  
    s = np.asarray(s_array_cm, dtype=float)

    #Initialising a DM array
    tot_S_nu_array = []
  
    a = a_Rsun * R_sun_cgs                                #Converts the semi-mmajor axis from solar radii to cm
    R_star = R_star_sol * R_sun_cgs                       #Converts the stellar radius from solar radii to cm
    Mdot = 10**Mdot_Msun_per_yr * (M_sun_cgs / yr_to_sec) #Converts the stellar mass loss from solar mass per year to g/s
    v_0 = v_0_kms * kms_cgs                               #Converts the stellar surface wind velocity ftom km/s to cm/s
    v_inf = v_inf_kms * kms_cgs                           #Converts the terminal wind velocity from km/s to cm/s
    k = Mdot / (4*np.pi*mu_e*m_p_cgs)                     #Calculates the k constant (cm^-1)
    M_star = M_star_sol * M_sun_cgs                       #Converts the stellar mass from solar mass to g

    for value in phi:
        for freq in nu:
            #Stellar wind number density
            r = a*(1 - e**2)/(1 + e*np.cos(value))                                                                                   #Calculates the orbital separation along the orbit
            r_s_sqrd = r**2 + 2*r*s*(np.cos(value+omega)*np.sin(i)*np.cos(Omega)+np.sin(value+omega)*np.sin(i)*np.sin(Omega)) + s**2 #Calculates the squared radial distance from the star along the line of sight
            v_r_s = v_0 + (v_inf-v_0)*(1-R_star/np.sqrt(r_s_sqrd))**beta                                                             #Calculates the velocity of the wind 
            n_e = k/(r_s_sqrd*v_r_s)                                                                                                 #Calculates the number density
            n_e[np.sqrt(r_s_sqrd) <= R_star] = 0                                                                                     #Eclipse condition
    
            #Disc number density
            r_disc = np.sqrt(r_s_sqrd)                                                                                                 #Projected radius through disc plane   
            c_s = np.sqrt((k_B_cgs * T_disc)/(mu_e * m_p_cgs))                                                                         #Calculates the disc speed of sound
            H_rdisc = c_s * np.sqrt(r_disc/(G_cgs * M_star)) * r_disc                                                                  #Calculates the scale height
            z = r*np.sin(value+omega)*np.sin(i_disc) + s*(np.sin(i+Omega)*np.sin(Omega)*np.sin(i_disc) + np.cos(i+Omega)*np.cos(i_disc)) #Calculates the height above the disc plane
            rho_disc = 10**rho0 * (r_disc/R_star)**(-beta_disc) * np.exp(-z**2/H_rdisc**2)                                             #Calculates the disc desnity profile
            ne_disc = rho_disc/(mu_e * m_p_cgs)                                                                                        #Calculates the number density
            ne_disc[r_disc < R_star] = 0                                                                                               #Constraining the inner disc radius

            #Total number density
            total_ne = n_e + ne_disc

            total_tau = 8.24e-2*freq**(-2.1)*((T_wind+T_disc)/2)**(-1.35)*np.trapz(total_ne**2, s / pc_to_cm) #Integrating the DM in cm^-2
            total_S_nu = freq**(-p)*np.exp(-total_tau)                                                        #Calculatinng the flux density
            tot_S_nu_array.append(total_S_nu)
        
    tot_S_nu_2D = np.array(tot_S_nu_array).reshape(len(phi), len(nu))                                         #Reshaping to a 2D array
    return np.array(tot_S_nu_2D) 
```

```python
#Total rotation measure
def tot_RM(phi, i, a_Rsun, R_star_sol, M_star_sol, Mdot_Msun_per_yr, v_0_kms, v_inf_kms, e, omega, Omega, beta, T_disc, beta_disc, rho0, i_disc, mu_e=1.2, s_array_cm=None):
    '''
    This function calculates the rotation measure of the signal through the stellar wind and circumstellar disc.
    ------------------------------------------------------------------------------------------------------------
    Parameters:

    phi              [float] -> Orbital Phase (rad)
    i                [float] -> Inclination (rad)
    a_Rsun           [float] -> Orbital radius (solar radii)
    R_star_sol       [float] -> Stellar radius (solar radii)
    M_star_sol       [float] -> Stellar mass (solar masses)
    Mdot_Msun_per_yr [float] -> Log wind mass loss (solar masses per year)
    v_0_kms          [float] -> Wind velocity at stellar radius (km/s)
    v_inf_kms        [float] -> Terminal velocity of wind (km/s)
    e                [float] -> Eccentricity (0 < e < 1)
    omega            [float] -> Argument of periastron (rad)
    Omega            [float] -> Longitude of ascending node (rad)
    beta             [float] -> Wind acceleration exponent
    T_disc           [float] -> Temperature of disc (K)
    beta_disc        [floar] -> Disc power-law exponent
    rho0             [float] -> Log disc density at stellar surface (gcm^-3)
    i_disc           [float] -> Inclination of disc
    mu_e             [float] -> Mean molecular weight per free electron
    s_array_cm       [array] -> Distance along line of sight from signal (AU)
    
    -----------------------------------------------------------------------------------------------------------------------------
    Returns:
    
    The function returns the total rotation measure of the signal [array] -> (rad cm^-2).
    '''
    if s_array_cm is None:
        s_array_cm = np.logspace(np.log10(1e5), np.log10(12 * AU_cgs), 512)  
    s = np.asarray(s_array_cm, dtype=float)

    #Initialising an RM array
    tot_RM_array = []
  
    a = a_Rsun * R_sun_cgs                                #Converts the semi-mmajor axis from solar radii to cm
    R_star = R_star_sol * R_sun_cgs                       #Converts the stellar radius from solar radii to cm
    Mdot = 10**Mdot_Msun_per_yr * (M_sun_cgs / yr_to_sec) #Converts the stellar mass loss from solar mass per year to g/s
    v_0 = v_0_kms * kms_cgs                               #Converts the stellar surface wind velocity ftom km/s to cm/s
    v_inf = v_inf_kms * kms_cgs                           #Converts the terminal wind velocity from km/s to cm/s
    k = Mdot / (4*np.pi*mu_e*m_p_cgs)                     #Calculates the k constant (cm^-1)
    M_star = (M_star_sol * M_sun_cgs)                     #Converts the stellar mass from solar mass to g
    B0 = 0.8                                              #Magnetic field at the stellar surface in G

    for value in phi:
        #Stellar wind number density
        r = a*(1 - e**2)/(1 + e*np.cos(value))                                                                                   #Calculates the orbital separation along the orbit
        r_s_sqrd = r**2 + 2*r*s*(np.cos(value+omega)*np.sin(i)*np.cos(Omega)+np.sin(value+omega)*np.sin(i)*np.sin(Omega)) + s**2 #Calculates the squared radial distance from the star along the line of sight
        v_r_s = v_0 + (v_inf-v_0)*(1-R_star/np.sqrt(r_s_sqrd))**beta                                                             #Calculates the velocity of the wind 
        n_e = k/(r_s_sqrd*v_r_s)                                                                                                 #Calculates the number density
        n_e[np.sqrt(r_s_sqrd) <= R_star] = 0                                                                                     #Eclipse condition
    
        #Disc number density
        r_disc = np.sqrt(r_s_sqrd)                                                                                                 #Projected radius through disc plane   
        c_s = np.sqrt((k_B_cgs * T_disc)/(mu_e * m_p_cgs))                                                                         #Calculates the disc speed of sound
        H_rdisc = c_s * np.sqrt(r_disc/(G_cgs * M_star)) * r_disc                                                                  #Calculates the scale height
        z = r*np.sin(value+omega)*np.sin(i_disc) + s*(np.sin(i+Omega)*np.sin(Omega)*np.sin(i_disc) + np.cos(i+Omega)*np.cos(i_disc)) #Calculates the height above the disc plane
        rho_disc = 10**rho0 * (r_disc/R_star)**(-beta_disc) * np.exp(-z**2/H_rdisc**2)                                             #Calculates the disc desnity profile
        ne_disc = rho_disc/(mu_e * m_p_cgs)                                                                                        #Calculates the number density
        ne_disc[r_disc < R_star] = 0                                                                                               #Constraining the inner disc radius

        #Total number density
        total_ne = n_e + ne_disc

        #Wind magnetic field
        B = B0 * (R_star/np.sqrt(r_s_sqrd))                                                                          #Total magnetic field
        B_parallel = B * (np.sin(i)*np.cos(Omega)*np.cos(value+omega) + np.sin(i)*np.sin(Omega)*np.sin(value+omega)) #Magnetic field parallel to line of sight
        RM_integrand = total_ne*B_parallel

        total_RM_cgs = 8.1e5 * np.trapz(RM_integrand, s / pc_to_cm)  #Integrating the RM in rad m^-2
        tot_RM_array.append(total_RM_cgs)

    return np.array(tot_RM_array) 
```

```python
#Curve_fit DM function
def DM_curve_fit(phi, i_orb, Mdot, omega, Omega, rho0, i_disc, e):
    '''
    This function employs scipy's curve_fit by implementing the general tot_DM function.
    -------------------------------------------------------------------------------------
    Parameters:

    phi    [array] -> Orbital Phase (rad)
    i_orb  [float] -> Orbital Inclination (rad)
    Mdot   [float] -> Log stellar Mass Loss (solar mass per year)
    e      [float] -> Eccentricity
    omega  [float] -> Argument of Periastron (rad)
    Omega  [float] -> Longitude of Ascending Node (rad)
    rho0   [float] -> Log initial Disc Density (gcm^-3)
    i_disc [float] -> Disc Inclination (rad)
    --------------------------------------------------------------------------------------
    Returns:

    The dipersion measure curve fit.
    '''
    #Fixing the remaining parameters of the model
    a_Rsun_fixed = 1091
    R_star_sol_fixed = 6
    M_star_sol_fixed = 10
    v_0_kms_fixed = 5
    v_inf_kms_fixed = 1e3
    beta_fixed = 2
    T_disc_fixed = 1.2e4
    beta_disc_fixed = 4

    return tot_DM(phi, i_orb, a_Rsun_fixed, R_star_sol_fixed, M_star_sol_fixed, Mdot, v_0_kms_fixed, v_inf_kms_fixed, e, omega, Omega, beta_fixed, T_disc_fixed, beta_disc_fixed, rho0, i_disc)
```

When plotting the number density as a function of line of sight distance, it is important to revisit the squared radial distance from the star along the line of sight:
## $r(s)^2 = (r_s + s\hat{n})^2 = r^2 + 2rs [cos(\phi + \omega) sin (i)cos(\Omega) + sin(\phi + \omega)sin(i)sin(\Omega)] + s^2$
As we can see, when
## $r >> s$, $r(s)^2 \approx r^2$
so
## $n_e(r(s)) \approx \frac{k}{r^2} = const$
Therefore, we can only start seeing a power law relation when 
## $r \sim s$

# Be Disc

## Be Disc Mass
As a preliminary step of analysing the Be disc it is important to check its total mass.

Using the definition of density
## $\rho = \frac{dM}{dV} \to dM = \rho dV \to M = \int \rho dV$

With
* $M$ = Disc mass
* $V$ = Disc volume

For discs, $dV = rdrd\phi dz$
With
* $dr$ = Radial change
* $d\phi$ = Angular change
* $dz$ = Vertical change

Since $\int_0^{2\pi} rd\phi = 2\pi r$, we end up with

## $M = 2\pi \int_{R_*}^{R_{out}} \int_{-\infty}^{\infty} \rho_0 (\frac{r}{R_*})^{-\beta} e^{\frac{-z^2}{H_{r}^2}} dzdr$
With
* $R_*$ = Stellar radius
* $R_{out}$ = Outer radius of disc

Using the gaussian integral $\int_{-\infty}^{\infty} e^{-ax^2} = \sqrt{\frac{\pi}{a}}$, the integral of the vertical change becomes

## $\int_{-\infty}^{\infty} e^{\frac{-z^2}{H_{r}^2}}dz = \sqrt{\pi}H_{r}$
The final expression of the disc mass is

## $M = 2\pi^{1.5} \rho_0 \int_{R_*}^{R_out} (\frac{r}{R_*})^{-\beta} H_r r dr$

```python
#Disc mass function
def discmass(R_star_sol, R_out_sol, M_star_sol, T_disc, beta_disc, rho0, mu_e=1.2):
    ''' This function calculates the mass of a circumstellar disc.
    ----------------------------------------------------------------
    Parameters:

    R_star_sol [float] -> #Stellar radius in solar radii (cm)
    R_out_sol  [float] -> #Disc outer radius in solar radii (cm)
    M_star_sol [float] -> #Stellar mass in solar mass (g)
    T_disc     [float] -> #Disc temperature (K)
    beta_disc  [float] -> #Disc density power-law exponent
    rho0       [float] -> #Disc density at stellar surface (g cm^-3)
    mu_e       [float] -> #Mean molecular weight per free electron
    -----------------------------------------------------------------
    Returns:

    The circumstellar disc mass in solar masses [float]
    '''


    R_star = R_star_sol * R_sun_cgs
    R_out = R_out_sol * R_sun_cgs
    M_star = M_star_sol * M_sun_cgs
    
    r = np.linspace(R_star, R_out, 512)

    c_s = np.sqrt((k_B_cgs*T_disc)/(mu_e*m_p_cgs))        #Speed of sound through disc (cm/s)

    H_r = c_s*np.sqrt(r/(G_cgs*M_star))*r                 #Disc scale height (cm)  

    integrand = (r/R_star)**(-beta_disc)*H_r*r            #Disc mass integrand

    M_disc_cgs = 2*np.pi**1.5*rho0*np.trapz(integrand, r) #Integrating the disc mass (g)

    M_disc = M_disc_cgs/M_sun_cgs                         #Disc mass in solar mass

    return M_disc
```

# Parameter Studies

```python
#DM function for parameter studies
def tot_DM_map(phi, i, i_disc, R_star_sol, M_star_sol, Mdot_Msun_per_yr, v_0_kms, v_inf_kms, e, omega, Omega, beta, T_disc, beta_disc, rho0, mu_e=1.2, s_array_cm=None):
    #Instantiating a line of sight array
    if s_array_cm is None:
        s_array_cm = np.logspace(np.log10(1e5), np.log10(12 * AU_cgs), 512)  
    s = np.asarray(s_array_cm, dtype=float)
    
    M_star = M_star_sol * M_sun_cgs                                   #Converts the stellar mass from solar mass to g
    P = 1.068e8                                                       #Orbital period in seconds (1237 days)
    a = ((P**2*G_cgs*(M_star_sol+1.4)*M_sun_cgs)/(4*np.pi**2))**(1/3) #Semi-mmajor axis (cm)
    R_star = R_star_sol * R_sun_cgs                                   #Converts the stellar radius from solar radii to cm
    Mdot = 10**Mdot_Msun_per_yr * (M_sun_cgs / yr_to_sec)             #Converts the stellar mass loss from solar mass per year to g/s
    v_0 = v_0_kms * kms_cgs                                           #Converts the stellar surface wind velocity ftom km/s to cm/s
    v_inf = v_inf_kms * kms_cgs                                       #Converts the terminal wind velocity from km/s to cm/s
    k = Mdot / (4*np.pi*mu_e*m_p_cgs)                                 #Calculates the k constant (cm^-1)

    #Stellar wind number density
    r = a*(1 - e**2)/(1 + e*np.cos(phi))                                                                                 #Calculates the orbital separation along the orbit
    r_s_sqrd = r**2 + 2*r*s*(np.cos(phi+omega)*np.sin(i)*np.cos(Omega)+np.sin(phi+omega)*np.sin(i)*np.sin(Omega)) + s**2 #Calculates the squared radial distance from the star along the line of sight
    v_r_s = v_0 + (v_inf-v_0)*(1-R_star/np.sqrt(r_s_sqrd))**beta                                                         #Calculates the velocity of the wind 
    n_e = k/(r_s_sqrd*v_r_s)                                                                                             #Calculates the number density
    n_e[np.sqrt(r_s_sqrd) <= R_star] = 0                                                                                 #Eclipse condition
    
    #Disc number density
    r_disc = np.sqrt(r_s_sqrd)                                                     #Projected radius through disc plane   
    c_s = np.sqrt((k_B_cgs * T_disc)/(mu_e * m_p_cgs))                             #Calculates the disc speed of sound
    H_rdisc = c_s * np.sqrt(r_disc/(G_cgs * M_star)) * r_disc                      #Calculates the scale height
    z = (r*np.sin(phi+omega)*np.sin(i_disc) + s*np.cos(i)*np.cos(i_disc))          #Calculates the height above the disc plane
    rho_disc = 10**rho0 * (r_disc/R_star)**(-beta_disc) * np.exp(-z**2/H_rdisc**2) #Calculates the disc desnity profile
    ne_disc = rho_disc/(mu_e * m_p_cgs)                                            #Calculates the number density
    ne_disc[r_disc < R_star] = 0                                                   #Constraining the inner disc radius

    #Total number density
    total_ne = n_e + ne_disc
    
    total_DM = np.trapz(total_ne, s) / pc_to_cm #Integrating the DM 

    return total_DM 
```

```python
#Free-free absorption function for parameter studies
def tot_ffa_map(phi, i, i_disc, R_star_sol, M_star_sol, Mdot_Msun_per_yr, v_0_kms, v_inf_kms, e, omega, Omega, beta, T_wind, T_disc, beta_disc, rho0, nu, mu_e=1.2, s_array_cm=None):
    #Instantiating a line of sight array
    if s_array_cm is None:
        s_array_cm = np.logspace(np.log10(1e5), np.log10(12 * AU_cgs), 512)  
    s = np.asarray(s_array_cm, dtype=float)
  
    M_star = M_star_sol * M_sun_cgs                                   #Converts the stellar mass from solar mass to g
    P = 1.068e8                                                       #Orbital period in seconds (1237 days)
    a = ((P**2*G_cgs*(M_star_sol+1.4)*M_sun_cgs)/(4*np.pi**2))**(1/3) #Semi-mmajor axis (cm)
    R_star = R_star_sol * R_sun_cgs                                   #Converts the stellar radius from solar radii to cm
    Mdot = 10**Mdot_Msun_per_yr * (M_sun_cgs / yr_to_sec)             #Converts the stellar mass loss from solar mass per year to g/s
    v_0 = v_0_kms * kms_cgs                                           #Converts the stellar surface wind velocity ftom km/s to cm/s
    v_inf = v_inf_kms * kms_cgs                                       #Converts the terminal wind velocity from km/s to cm/s
    k = Mdot / (4*np.pi*mu_e*m_p_cgs)                                 #Calculates the k constant (cm^-1)

    #Stellar wind number density
    r = a*(1 - e**2)/(1 + e*np.cos(phi))                                                                                 #Calculates the orbital separation along the orbit
    r_s_sqrd = r**2 + 2*r*s*(np.cos(phi+omega)*np.sin(i)*np.cos(Omega)+np.sin(phi+omega)*np.sin(i)*np.sin(Omega)) + s**2 #Calculates the squared radial distance from the star along the line of sight
    v_r_s = v_0 + (v_inf-v_0)*(1-R_star/np.sqrt(r_s_sqrd))**beta                                                         #Calculates the velocity of the wind 
    n_e = k/(r_s_sqrd*v_r_s)                                                                                             #Calculates the number density
    n_e[np.sqrt(r_s_sqrd) <= R_star] = 0                                                                                 #Eclipse condition
    
    #Disc number density
    r_disc = np.sqrt(r_s_sqrd)                                                     #Projected radius through disc plane   
    c_s = np.sqrt((k_B_cgs * T_disc)/(mu_e * m_p_cgs))                             #Calculates the disc speed of sound
    H_rdisc = c_s * np.sqrt(r_disc/(G_cgs * M_star)) * r_disc                      #Calculates the scale height
    z = (r*np.sin(phi+omega)*np.sin(i_disc) + s*np.cos(i)*np.cos(i_disc))          #Calculates the height above the disc plane
    rho_disc = 10**rho0 * (r_disc/R_star)**(-beta_disc) * np.exp(-z**2/H_rdisc**2) #Calculates the disc desnity profile
    ne_disc = rho_disc/(mu_e * m_p_cgs)                                            #Calculates the number density
    ne_disc[r_disc < R_star] = 0                                                   #Constraining the inner disc radius

    #Total number density
    total_ne = n_e + ne_disc

    total_tau = 8.24e-2*nu**(-2.1)*((T_wind+T_disc)/2)**(-1.35)*np.trapz(total_ne**2, s / pc_to_cm) #Integrating the optical depth

    return total_tau 
```
